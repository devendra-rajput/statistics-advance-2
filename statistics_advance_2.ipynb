1. What is hypothesis testing in statistics?
   - Hypothesis testing is a formal statistical method used to make decisions or inferences about population parameters based on sample data. It's commonly used to test whether a claim or assumption (called a hypothesis) about a population is true.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
   - The null hypothesis (H₀) is a statement that assumes there is no effect or difference—it represents the status quo. The alternative hypothesis (H₁) suggests there is an effect or difference. Hypothesis testing evaluates whether there's enough evidence to reject H₀ in favor of H₁. If the evidence is strong, we support the alternative hypothesis.

3.  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
    - The significance level (denoted as α) is the threshold used to decide whether to reject the null hypothesis. It represents the probability of making a Type I error, which means rejecting a true null hypothesis. Common values are 0.05 or 0.01. If the p-value ≤ α, we reject the null hypothesis; if p-value > α, we fail to reject it.

4. What are Type I and Type II errors? Give examples of each.
    *   A Type I error occurs when the null hypothesis is rejected when it's actually true (false positive). A Type II error happens when the null hypothesis is not rejected when it's actually false (false negative).
      **Example:** Type I - convicting an innocent person; Type II - letting a guilty person go free.

5. What is the difference between a Z-test and a T-test? Explain when to use
each?
   - A Z-test is used when the population standard deviation is known and the sample size is large (n ≥ 30). A T-test is used when the population standard deviation is unknown and the sample size is small (n < 30).
      Use a Z-test for large samples or known variance, and a T-test for small samples or unknown variance.

```python
"""
Q-7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

      sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                      50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
"""

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                      50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

population_std = 0.5

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

mu_0 = 50

z_stat = (sample_mean - mu_0) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis (H₀).")
else:
    print("Fail to reject the null hypothesis (H₀).")
```

```python
# Q-8.  Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
mu = 100
sigma = 15
n = 100

data = np.random.normal(mu, sigma, n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
standard_error = sample_std / np.sqrt(n)

confidence_level = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence_level)/2)
margin_of_error = z_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')

plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Q-9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):

    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=100)
z_scores = calculate_and_plot_z_scores(sample_data)

```
